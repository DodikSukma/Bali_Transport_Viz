```python
tahun = dataset['TAHUN']
tahun.head(2)
```

```python
bus = dataset['BUS']
sm = dataset['SEPEDA MOTOR']
truck = dataset['TRUK']
mp = dataset['MOBIL PENUMPANG']

```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Bar(x=tahun,
                y=bus,
                name='BUS',
                marker_color='rgb(55, 83, 109)'
                ))
fig.add_trace(go.Bar(x=tahun,
                y=sm,
                name='SEPEDA MOTOR',
                marker_color='rgb(26, 118, 255)'
                ))
fig.add_trace(go.Bar(x=tahun,
                y=truck,
                name='TRUCK',
                marker_color='rgb(26, 118, 255)'
                ))
fig.add_trace(go.Bar(x=tahun,
                y=mp,
                name='MOBIL PENUMPANG',
                marker_color='rgb(26, 118, 255)'
                ))

fig.update_layout(
    title='JUMLAH KENDARAAN DI PROVINSI BALI DARI TAHUN KE TAHUN',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='JUMLAH',
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=0,
        y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    barmode='group',
    bargap=0.15, # gap between bars of adjacent location coordinates.
    bargroupgap=0.1 # gap between bars of the same location coordinate.
)
fig.show()
```
